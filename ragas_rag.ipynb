```python
# ================================================
# Step 2. 🔐 API Key Setup (Securely)
# ================================================
import getpass

# Use getpass to safely input your Groq API key in Colab
groq_api_key = getpass.getpass("Enter your Groq API key: ")

print("✅ Groq API key securely loaded.")
```

```python
# ================================================
# Step 3. 📦 Imports for RAG and Evaluation
# ================================================
import os
import pandas as pd
from groq import Groq
from langchain_community.vectorstores import FAISS
from langchain.docstore.document import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.base import Embeddings
from sentence_transformers import SentenceTransformer
from datasets import Dataset
from langchain_groq import ChatGroq

# RAGAS evaluation metrics
from ragas import evaluate
from ragas.metrics import faithfulness, context_precision, context_recall, answer_correctness
```

```python
# ================================================
# Step 4. 🧬 Define Local Embeddings Class
# ================================================
class LocalEmbeddings(Embeddings):
    def __init__(self, model_name="sentence-transformers/all-mpnet-base-v2"):
        self.model = SentenceTransformer(model_name)

    def embed_documents(self, texts):
        return self.model.encode(texts, show_progress_bar=False).tolist()

    def embed_query(self, text):
        return self.model.encode([text], show_progress_bar=False)[0].tolist()

# Initialize embeddings
embeddings = LocalEmbeddings()
```

```python
# ================================================
# Step 5. ⚙️ Initialize Groq LLM
# ================================================
client = Groq(api_key=groq_api_key)
```

```python
# ================================================
# Step 6. 🧱 Prepare Knowledge Base
# ================================================
texts = [
    "The Taj Mahal is a white marble mausoleum located in Agra, built by Mughal Emperor Shah Jahan in memory of his wife Mumtaz Mahal.",
    "The Qutub Minar is a UNESCO World Heritage site in Delhi, built in 1193 by Qutb-ud-din Aibak.",
    "The Red Fort in Delhi was the main residence of the Mughal emperors for nearly 200 years, until 1857.",
    "Hampi is an ancient village in Karnataka, known for the ruins of the Vijayanagara Empire.",
    "The Sun Temple in Konark, Odisha, is designed in the shape of a colossal chariot.",
    "The Gateway of India is an arch-monument built in the early 20th century in Mumbai.",
    "Mysore Palace is a historical palace and a royal residence in the city of Mysore, Karnataka."
]

docs = [Document(page_content=t) for t in texts]
splitter = RecursiveCharacterTextSplitter(chunk_size=250, chunk_overlap=30)
chunks = splitter.split_documents(docs)

# Create FAISS vector store
db = FAISS.from_documents(chunks, embeddings)
retriever = db.as_retriever(search_kwargs={"k": 3})
```

```python
# ================================================
# Step 7. 🧠 Define the RAG Pipeline
# ================================================
def rag_pipeline(query, retriever, client, model="llama-3.1-8b-instant"):
    # Retrieve relevant context
    retrieved_docs = retriever.invoke(query)
    context = "\n".join([doc.page_content for doc in retrieved_docs])

    system_prompt = f"""
You are a helpful AI assistant. Use ONLY the context below to answer accurately.
If the answer is not found, respond with "I don't know".

Context:
{context}
    """

    chat_completion = client.chat.completions.create(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": query}
        ],
        model=model,
        temperature=0.1,
        max_tokens=512
    )

    answer = chat_completion.choices[0].message.content
    contexts = [doc.page_content for doc in retrieved_docs]

    return {"question": query, "answer": answer, "contexts": contexts}
```

```python
# ================================================
# Step 8. 💬 Define Queries and Ground Truths
# ================================================
queries = [
    "Where is the Taj Mahal located?",
    "Who built the Qutub Minar?",
    "What is special about Hampi?",
    "Which city has the Gateway of India?",
    "Describe the Sun Temple in Konark."
]

ground_truths = [
    "The Taj Mahal is located in Agra.",
    "The Qutub Minar was built by Qutb-ud-din Aibak.",
    "Hampi is known for the ruins of the Vijayanagara Empire in Karnataka.",
    "The Gateway of India is located in Mumbai.",
    "The Sun Temple in Konark, Odisha, is designed like a colossal chariot."
]
```

```python
# ================================================
# Step 9. 🚀 Run RAG Pipeline
# ================================================
results = []
for query in queries:
    result = rag_pipeline(query, retriever, client)
    results.append(result)
    print(f"✅ Q: {query}\n🧩 A: {result['answer']}\n")
```

```python
# ================================================
# Step 10. 🧾 Prepare Data for RAGAS Evaluation
# ================================================
data = {
    "question": [r["question"] for r in results],
    "contexts": [r["contexts"] for r in results],
    "answer": [r["answer"] for r in results],
    "ground_truth": ground_truths
}

dataset = Dataset.from_dict(data)
```

```python
# ================================================
# Step 11. 📊 Evaluate with RAGAS Metrics
# ================================================
llm_eval = ChatGroq(
    groq_api_key=groq_api_key,
    model_name="llama-3.3-70b-versatile"
)

scores = evaluate(
    dataset=dataset,
    metrics=[faithfulness, answer_correctness, context_precision, context_recall],
    llm=llm_eval,
    embeddings=embeddings,
    raise_exceptions=False,
)
```

```python
# ================================================
# Step 12. 📈 Display Results
# ================================================
results_df = pd.DataFrame({
    "Question": data["question"],
    "Answer": data["answer"],
    "Faithfulness": scores["faithfulness"],
    "Answer Correctness": scores["answer_correctness"],
    "Context Precision": scores["context_precision"],
    "Context Recall": scores["context_recall"]
})

print("\n=== 🧾 RAGAS Evaluation Results ===")
print(results_df)
print("\n=== 🔍 Average Scores ===")
print(f"Faithfulness: {scores['faithfulness'][0]:.3f}")
print(f"Answer Correctness: {scores['answer_correctness'][0]:.3f}")
print(f"Context Precision: {scores['context_precision'][0]:.3f}")
print(f"Context Recall: {scores['context_recall'][0]:.3f}")
```
